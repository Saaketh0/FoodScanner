```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd
import numpy as np
from pathlib import Path
from datasets import load_from_disk
import torch
loaded_dataset = load_from_disk('embeddings/total_dataset')
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
df = loaded_dataset.select(range(3729280)).to_pandas()
df.shape
```

```python
embeds = df['embeddings'].to_numpy()
embeds = np.stack(embeds).astype(np.float16)
embeds.shape
```

```python
np.save('embeddings/embeddings_only.npy', embeds)
print("\nArray saved to my_array.npy")
```

```python
# 3. Load the array back from the file
loaded_array = np.load('embeddings/embeddings_only.npy')
print("\n--- Loaded Array ---")
print(loaded_array.shape)
```

```python
sentences = ["Peanut Butter"]
embeddings = np.array(model.encode(sentences), dtype=np.float16)
cosine_scores = util.cos_sim(embeddings, embeds)
print(str(int(100*(torch.max(cosine_scores)).item())) + "% similarity")  # Output: 9
max_index = (torch.argmax(cosine_scores)).item()
print(str(max_index) + " index")
print(loaded_dataset["product_name"][max_index])
print(loaded_dataset["nutriments"][max_index][0])
```
